```python
target = df["Price"]
features = df.drop("Price", axis = 1)
ohe_features = pd.get_dummies(features, drop_first=True)
train_val_features, test_features, train_val_target, test_target = train_test_split(features, target, test_size=0.2, random_state = 123453)
train_features, val_features, train_target, val_target = train_test_split(train_val_features, train_val_target, test_size=0.25, random_state = 123453)
ohe_train_features = ohe_features.loc[train_features.index, :]
ohe_test_features = ohe_features.loc[test_features.index, :]
ohe_val_features = ohe_features.loc[val_features.index, :]

```

## LightGBM

```python
%%time
model = LGBMRegressor()
learning_rate = np.logspace(-1, 0, 5)
num_leaves = np.logspace(1, 3, 5).astype(int)
params = [{'num_leaves': num_leaves,
            'learning_rate': learning_rate,
            'random_state': [12345]
           }]

clf = GridSearchCV(model, params, scoring= "neg_mean_squared_error")
clf.fit(ohe_train_features, train_target)

print(f"Best params: {clf.best_params_}")
best_rmse = np.sqrt(-clf.best_score_)
print(f"Best RMSE: {best_rmse:.6f}")
print("-" * 50)



results = []
for i in range(len(clf.cv_results_['params'])):
    rmse = np.sqrt(-clf.cv_results_['mean_test_score'][i])
    std = np.sqrt(clf.cv_results_['std_test_score'][i])  
    results.append((rmse, std, clf.cv_results_['params'][i]))

results.sort(key=lambda x: x[0])

print("All results sorted by RMSE:")
for i, (rmse, std, params) in enumerate(results, 1):
    print(f"{i}. RMSE: {rmse:.6f} ± {std:.6f} | Params: {params}")
```

```python
%%time
regressor = LGBMRegressor(learning_rate=0.1778279410038923, num_leaves=316, random_state=12345)
regressor.fit(ohe_train_features, train_target)
prediction = regressor.predict(ohe_test_features)
prediction1 = regressor.predict(ohe_train_features)
rmse_lighGBM = root_mean_squared_error(test_target, prediction)
rmse1 = root_mean_squared_error(train_target, prediction1)
print(f"RMSE: {rmse_lighGBM}")
print(f"RMSE: {rmse1}")
```

## Линейная регрессия 

```python
%%time
model = LinearRegression()
cvdf_rmse = cross_val_score(model, ohe_train_features, train_target,
                          cv=4, scoring='neg_mean_squared_error') 
rmse_linear_reg = (cvdf_rmse.mean() * -1) ** 0.5

print(f"RMSE: {rmse_linear_reg}")
```

```python
%%time
model = RandomForestRegressor(max_depth = 10, n_estimators = 100, random_state = 12345)
model.fit(ohe_train_features, train_target)
prediction = model.predict(ohe_test_features)
prediction1 = model.predict(ohe_train_features)
rmse_radnom_forest = root_mean_squared_error(test_target, prediction)
rmse1 = root_mean_squared_error(train_target, prediction1)
print(f"RMSE test: {rmse_radnom_forest}")
print(f"RMSE train: {rmse1}")
```
