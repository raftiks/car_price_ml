```python
target = df["LogPrice"]
features = df.drop(["Price", "LogPrice"], axis = 1)

categorical_cols = features.select_dtypes(include=["object"]).columns
encoder = OneHotEncoder(handle_unknown="ignore",sparse_output=False)
ohe_features = pd.DataFrame(
    encoder.fit_transform(features[categorical_cols]),
    columns=encoder.get_feature_names_out(categorical_cols),
    index=features.index
)

# Объединяем числовые и категориальные признаки
numeric_features = features.drop(columns=categorical_cols)
final_features = pd.concat([numeric_features, ohe_features], axis=1)

train_val_features, test_features, train_val_target, test_target = train_test_split(final_features, target, test_size=0.15, random_state = 123453)
train_features, val_features, train_target, val_target = train_test_split(train_val_features, train_val_target, test_size=0.2, random_state = 123453)




```

## LightGBM

```python
%%time
model = LGBMRegressor(verbose = 0)
learning_rate = np.logspace(-1, 0, 5)
num_leaves = np.logspace(1, 3, 5).astype(int)
params = [{'num_leaves': num_leaves,
            'learning_rate': learning_rate,
            'random_state': [12345]
            }]

clf = GridSearchCV(model, params, scoring= "neg_mean_squared_error")
clf.fit(train_features, train_target)

print(f"Best params: {clf.best_params_}")
best_rmse = np.sqrt(-clf.best_score_)
print(f"Best RMSE: {best_rmse:.6f}")
print("-" * 50)



results = []
for i in range(len(clf.cv_results_['params'])):
    rmse = np.sqrt(-clf.cv_results_['mean_test_score'][i])
    std = np.sqrt(clf.cv_results_['std_test_score'][i])  
    results.append((rmse, std, clf.cv_results_['params'][i]))

results.sort(key=lambda x: x[0])

print("All results sorted by RMSE:")
for i, (rmse, std, params) in enumerate(results, 1):
    print(f"{i}. RMSE: {rmse:.6f} ± {std:.6f} | Params: {params}")
```

```python
%%time
regressor = LGBMRegressor(learning_rate=0.1778279410038923, num_leaves=316, random_state=12345)
regressor.fit(train_features, train_target)

prediction_lag_test = regressor.predict(val_features)
prediction_test = np.expm1(prediction_lag_test)

prediction1_lag_train = regressor.predict(train_features)
prediction_train =  np.expm1(prediction1_lag_train)


rmse_lighGBM_log = root_mean_squared_error(val_target, prediction_lag_test)
rmse_train_log = root_mean_squared_error(train_target, prediction1_lag_train)

rmse_lighGBM = root_mean_squared_error(np.expm1(val_target), prediction_test)
rmse_train = root_mean_squared_error(np.expm1(train_target), prediction_train)

print(f"RMSE_log {rmse_lighGBM_log}")
print(f"RMSE_log_train: {rmse_train_log}")
print("-"*30)
print(f"RMSE: {rmse_lighGBM}")
print(f"RMSE_train: {rmse_train}")

```

```python
%%time
learning_rate = np.logspace(-2, 0, 5)
model = CatBoostRegressor(
    iterations = 2500,
    loss_function='RMSE'
)
params = {
    'learning_rate': learning_rate,
}

clf = GridSearchCV(
    estimator=model,
    param_grid=params,
    scoring='neg_mean_squared_error'
)

clf.fit(
    train_features, 
    train_target,
    eval_set = (val_features, val_target),
    early_stopping_rounds=50,
    verbose = 0
)


print(f"Best params: {clf.best_params_}")
best_rmse = np.sqrt(-clf.best_score_)
print(f"Best RMSE: {best_rmse:.6f}")
print("-" * 50)


results = []
for i in range(len(clf.cv_results_['params'])):
    rmse = np.sqrt(-clf.cv_results_['mean_test_score'][i])
    std = np.sqrt(clf.cv_results_['std_test_score'][i])  
    results.append((rmse, std, clf.cv_results_['params'][i]))

results.sort(key=lambda x: x[0])

print("All results sorted by RMSE:")
for i, (rmse, std, params) in enumerate(results, 1):
    print(f"{i}. RMSE: {rmse:.6f} ± {std:.6f} | Params: {params}")
```

```python
%%time
model = CatBoostRegressor(
    iterations = 2500,
    loss_function='RMSE',
    random_state=123453,
    early_stopping_rounds=50,
    verbose = 0,
    learning_rate = 0.31622776601683794)
model.fit(train_features, 
    train_target)
prediction_lag_test = model.predict(test_features)
prediction_test = np.expm1(prediction_lag_test)

prediction1_lag_train = model.predict(train_features)
prediction_train =  np.expm1(prediction1_lag_train)


rmse_Catboost_log = root_mean_squared_error(test_target, prediction_lag_test)
rmse_train_log = root_mean_squared_error(train_target, prediction1_lag_train)

rmse_Catboost = root_mean_squared_error(np.expm1(test_target), prediction_test)
rmse_train = root_mean_squared_error(np.expm1(train_target), prediction_train)

print(f"RMSE_log {rmse_Catboost_log}")
print(f"RMSE_log_train: {rmse_train_log}")
print("-"*30)
print(f"RMSE: {rmse_Catboost}")
print(f"RMSE_train: {rmse_train}")
```

## Линейная регрессия 

```python
%%time
model = TransformedTargetRegressor(
    regressor=LinearRegression(),
    func= lambda y: y,       
    inverse_func=np.expm1  # exp(y)-1, чтобы вернуть к исходной шкале
)

y_pred_original = cross_val_predict(model, train_features, train_target, cv=4)

# RMSE в исходной шкале
rmse_lin_regress = root_mean_squared_error(np.expm1(train_target), y_pred_original)
print(rmse_lin_regress)

```

```python
%%time
model = RandomForestRegressor(max_depth = 15,min_samples_leaf=20 ,n_estimators = 30, random_state = 12345)
model.fit(train_features, train_target)

prediction_lag_test = model.predict(test_features)
prediction_test = np.expm1(prediction_lag_test)

prediction1_lag_train = model.predict(train_features)
prediction_train =  np.expm1(prediction1_lag_train)


rmse_forest_log = root_mean_squared_error(test_target, prediction_lag_test)
rmse_train_log = root_mean_squared_error(train_target, prediction1_lag_train)

rmse_forest = root_mean_squared_error(np.expm1(test_target), prediction_test)
rmse_train = root_mean_squared_error(np.expm1(train_target), prediction_train)
print(f"RMSE_log {rmse_forest_log}")
print(f"RMSE_log_train: {rmse_train_log}")
print("-"*30)
print(f"RMSE: {rmse_forest}")
print(f"RMSE_train: {rmse_train}")
```

```python
%%time
model = DecisionTreeRegressor(max_depth=15,min_samples_leaf=40,random_state = 12345)
model.fit(train_features, train_target)

prediction_lag_test = model.predict(test_features)
prediction_test = np.expm1(prediction_lag_test)

prediction1_lag_train = model.predict(train_features)
prediction_train =  np.expm1(prediction1_lag_train)


rmse_tree_log = root_mean_squared_error(test_target, prediction_lag_test)
rmse_train_log = root_mean_squared_error(train_target, prediction1_lag_train)

rmse_tree = root_mean_squared_error(np.expm1(test_target), prediction_test)
rmse_train = root_mean_squared_error(np.expm1(train_target), prediction_train)
print(f"RMSE_log {rmse_tree_log}")
print(f"RMSE_log_train: {rmse_train_log}")
print("-"*30)
print(f"RMSE: {rmse_tree}")
print(f"RMSE_train: {rmse_train}")
```

```python
%%time
model = TransformedTargetRegressor(
    regressor=Ridge(),
    func= lambda y: y,       
    inverse_func=np.expm1  # exp(y)-1, чтобы вернуть к исходной шкале
)

y_pred_original = cross_val_predict(model, train_features, train_target, cv=4)

# RMSE в исходной шкале
rmse_Ridge = root_mean_squared_error(np.expm1(train_target), y_pred_original)
print(rmse_Ridge)
```
